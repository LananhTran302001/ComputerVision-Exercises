# Implement an edge detector

Input:

```python
img = cv2.imread("../hanoi.jpg")
plt.imshow(img)
```

Step 1: Blur the image a little
\begin{equation*}
B_{\sigma} (x) = G_{\sigma} (x) * I(x)
\end{equation*}

```python
blur = cv2.GaussianBlur(img, (5, 5), 0)
```

Step 2: Contruct a Gaussian pyramid
\begin{equation*}
P = Pyramid( B_{\sigma} (x) )
\end{equation*}

```python
pyramid = []
layer = img.copy()
for i in range(4):
    layer = cv2.pyrDown(layer)
    pyramid.append(layer)
```

Step 3: Subtract an interpolated coarser-level pyramid image from the original resolution blurred image

\begin{equation*}
S(x) = B_{\sigma} (x) - P.InterpolatedLevel(L)
\end{equation*}

Step 4: For each quad of pixels, {(i, j),(i + 1, j),(i, j + 1),(i + 1, j + 1)}, count the number of zero crossings along the four edges.

Step 5: When there are exactly two zero crossings, compute their locations using (7.25) and store these edgel <br>
endpoints along with the midpoint in the edgel structure

Step 6: For each edgel, compute the local gradient by taking the horizontal and vertical differences between the <br>
values of S along the zero crossing edges

Step 7: Store the magnitude of this gradient as the edge strength and either its orientation or that of the segment <br>
joining the edgel endpoints as the edge orientation

Step 8: Add the edgel to a list of edgels or store it in a 2D array of edgels (addressed by pixel coordinates)
